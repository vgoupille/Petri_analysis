```python
import cv2
import numpy as np
import os

def clean_images_in_directory(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    for filename in os.listdir(input_dir):
        if filename.lower().endswith(".jpg"):
            input_path = os.path.join(input_dir, filename)

            # Charger image et passer en niveaux de gris
            image_rgb = cv2.cvtColor(cv2.imread(input_path), cv2.COLOR_BGR2RGB)
            image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

            # Binarisation
            _, image_thresh = cv2.threshold(image_gray, 128, 255, cv2.THRESH_BINARY)

            # Inverser pour trouver le contour externe noir
            image_inv = cv2.bitwise_not(image_thresh)
            contours, _ = cv2.findContours(image_inv, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

            if contours:
                # Supprimer le plus gros contour noir
                biggest_contour = max(contours, key=cv2.contourArea)
                mask = np.ones_like(image_thresh) * 255
                cv2.drawContours(mask, [biggest_contour], -1, 0, thickness=cv2.FILLED)

                # Ne garder que le plus grand objet blanc
                num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(mask, connectivity=8)
                areas = stats[1:, cv2.CC_STAT_AREA]
                if len(areas) > 0:
                    largest_label = 1 + np.argmax(areas)
                    mask_largest = np.zeros_like(mask)
                    mask_largest[labels == largest_label] = 255

                    # Appliquer le masque
                    mask_rgb = cv2.cvtColor(mask_largest, cv2.COLOR_GRAY2RGB)
                    image_masked = cv2.bitwise_and(image_rgb, mask_rgb)

                    # 🔥 Supprimer les bordures noires : crop à la bounding box du masque
                    coords = cv2.findNonZero(mask_largest)  # coordonnées des pixels blancs
                    x, y, w, h = cv2.boundingRect(coords)
                    image_cropped = image_masked[y:y+h, x:x+w]

                    # Sauvegarde en BGR
                    output_path = os.path.join(output_dir, filename)
                    cv2.imwrite(output_path, cv2.cvtColor(image_cropped, cv2.COLOR_RGB2BGR))
                    print(f"✅ Cleaned, cropped and saved: {output_path}")


# === Exemple d'utilisation ===
cleaned_dir = os.path.join(Abra_dir, "cleaned")
cropped_dir = os.path.join(Abra_dir)#, "test")
clean_images_in_directory(cropped_dir, cleaned_dir)
```

```python
import cv2
import numpy as np
import os

def split_objects_in_directory(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    for filename in os.listdir(input_dir):
        if filename.lower().endswith(".jpg"):
            input_path = os.path.join(input_dir, filename)

            # Charger image et passer en niveaux de gris
            image_rgb = cv2.cvtColor(cv2.imread(input_path), cv2.COLOR_BGR2RGB)
            image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

            # Binarisation
            _, image_thresh = cv2.threshold(image_gray, 128, 255, cv2.THRESH_BINARY)

            # Inverser pour trouver le contour externe noir
            image_inv = cv2.bitwise_not(image_thresh)
            contours, _ = cv2.findContours(image_inv, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

            if contours:
                # Supprimer le plus gros contour noir
                biggest_contour = max(contours, key=cv2.contourArea)
                mask = np.ones_like(image_thresh) * 255
                cv2.drawContours(mask, [biggest_contour], -1, 0, thickness=cv2.FILLED)

                # Trouver toutes les composantes blanches (objets d'intérêt)
                num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(mask, connectivity=8)

                for i in range(1, num_labels):  # ignorer le fond (0)
                    x, y, w, h, area = stats[i]
                    if area < 1000:  # 🔧 ignorer les petits bruits
                        continue

                    # Créer un masque pour cet objet
                    mask_obj = np.zeros_like(mask)
                    mask_obj[labels == i] = 255

                    # Appliquer le masque à l'image originale
                    mask_rgb = cv2.cvtColor(mask_obj, cv2.COLOR_GRAY2RGB)
                    image_masked = cv2.bitwise_and(image_rgb, mask_rgb)

                    # Crop à la bounding box
                    image_cropped = image_masked[y:y+h, x:x+w]

                    # Sauvegarde avec suffixe (_1, _2, …)
                    base, ext = os.path.splitext(filename)
                    output_path = os.path.join(output_dir, f"{base}_{i}{ext}")
                    cv2.imwrite(output_path, cv2.cvtColor(image_cropped, cv2.COLOR_RGB2BGR))

                    print(f"✅ Extracted object {i} from {filename} → {output_path}")


# === Exemple d'utilisation ===
cleaned_dir = os.path.join(Abra_dir, "objects")
cropped_dir = os.path.join(Abra_dir)
split_objects_in_directory(cropped_dir, cleaned_dir)
```

```python
import cv2
import numpy as np
import os

def is_round(mask_obj, threshold=0.75):
    # Trouver le plus grand contour de l'objet
    contours, _ = cv2.findContours(mask_obj, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return False
    
    cnt = max(contours, key=cv2.contourArea)
    area = cv2.contourArea(cnt)
    perimeter = cv2.arcLength(cnt, True)

    if perimeter == 0:
        return False

    # Circularité = 4π * Aire / (Périmètre²)
    circularity = 4 * np.pi * area / (perimeter ** 2)

    # Plus circularité → proche de 1 = cercle parfait
    return circularity > threshold


def split_objects_in_directory(input_dir, output_dir, threshold=0.75):
    os.makedirs(output_dir, exist_ok=True)

    for filename in os.listdir(input_dir):
        if filename.lower().endswith(".jpg"):
            input_path = os.path.join(input_dir, filename)

            # Charger image et passer en niveaux de gris
            image_rgb = cv2.cvtColor(cv2.imread(input_path), cv2.COLOR_BGR2RGB)
            image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

            # Binarisation
            _, image_thresh = cv2.threshold(image_gray, 128, 255, cv2.THRESH_BINARY)

            # Inverser pour trouver le contour externe noir
            image_inv = cv2.bitwise_not(image_thresh)
            contours, _ = cv2.findContours(image_inv, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

            if contours:
                # Supprimer le plus gros contour noir
                biggest_contour = max(contours, key=cv2.contourArea)
                mask = np.ones_like(image_thresh) * 255
                cv2.drawContours(mask, [biggest_contour], -1, 0, thickness=cv2.FILLED)

                # Trouver toutes les composantes blanches (objets d'intérêt)
                num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(mask, connectivity=8)

                for i in range(1, num_labels):  # ignorer le fond (0)
                    x, y, w, h, area = stats[i]
                    if area < 1000:  # ignorer petits bruits
                        continue

                    # Créer un masque pour cet objet
                    mask_obj = np.zeros_like(mask)
                    mask_obj[labels == i] = 255

                    # Vérifier si l'objet est rond
                    if not is_round(mask_obj, threshold=threshold):
                        continue

                    # Appliquer le masque à l'image originale
                    mask_rgb = cv2.cvtColor(mask_obj, cv2.COLOR_GRAY2RGB)
                    image_masked = cv2.bitwise_and(image_rgb, mask_rgb)

                    # Crop à la bounding box
                    image_cropped = image_masked[y:y+h, x:x+w]

                    # Sauvegarde avec suffixe (_1, _2, …)
                    base, ext = os.path.splitext(filename)
                    output_path = os.path.join(output_dir, f"{base}_{i}{ext}")
                    cv2.imwrite(output_path, cv2.cvtColor(image_cropped, cv2.COLOR_RGB2BGR))

                    print(f"✅ Extracted ROUND object {i} from {filename} → {output_path}")


# === Exemple d'utilisation ===
cleaned_dir = os.path.join(Abra_dir, "objects_cleaned4")
cropped_dir = os.path.join(Abra_dir)
split_objects_in_directory(cropped_dir, cleaned_dir, threshold=0.2)
```

```python
import cv2
import numpy as np
import os

def is_round_fast(mask_obj, tol=0.75):
    """Vérifie si un objet est rond en comparant aire et bounding box"""
    contours, _ = cv2.findContours(mask_obj, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return False
    
    cnt = max(contours, key=cv2.contourArea)
    area = cv2.contourArea(cnt)
    x, y, w, h = cv2.boundingRect(cnt)

    # Aire attendue pour un disque inscrit dans la bbox
    r = min(w, h) / 2.0
    circle_area = np.pi * (r ** 2)

    # Si l'objet a une aire proche de celle d'un disque → rond
    return area / circle_area > tol


def split_round_objects_in_directory(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)

    for filename in os.listdir(input_dir):
        if filename.lower().endswith(".jpg"):
            input_path = os.path.join(input_dir, filename)

            # Charger image et passer en niveaux de gris
            image_rgb = cv2.cvtColor(cv2.imread(input_path), cv2.COLOR_BGR2RGB)
            image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

            # Binarisation
            _, image_thresh = cv2.threshold(image_gray, 128, 255, cv2.THRESH_BINARY)

            # Inverser pour trouver le contour externe noir
            image_inv = cv2.bitwise_not(image_thresh)
            contours, _ = cv2.findContours(image_inv, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

            if contours:
                # Supprimer le plus gros contour noir
                biggest_contour = max(contours, key=cv2.contourArea)
                mask = np.ones_like(image_thresh) * 255
                cv2.drawContours(mask, [biggest_contour], -1, 0, thickness=cv2.FILLED)

                # Composantes = objets
                num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(mask, connectivity=8)

                for i in range(1, num_labels):
                    x, y, w, h, area = stats[i]
                    if area < 1000:
                        continue

                    # Créer un masque pour l’objet
                    mask_obj = np.zeros_like(mask)
                    mask_obj[labels == i] = 255

                    # Vérif "rond"
                    if not is_round_fast(mask_obj, tol=0.75):
                        print(f"❌ Objet {i} de {filename} rejeté (pas rond)")
                        continue

                    # Appliquer masque
                    mask_rgb = cv2.cvtColor(mask_obj, cv2.COLOR_GRAY2RGB)
                    image_masked = cv2.bitwise_and(image_rgb, mask_rgb)

                    # Crop
                    image_cropped = image_masked[y:y+h, x:x+w]

                    # Sauvegarde
                    base, ext = os.path.splitext(filename)
                    output_path = os.path.join(output_dir, f"{base}_round_{i}{ext}")
                    cv2.imwrite(output_path, cv2.cvtColor(image_cropped, cv2.COLOR_RGB2BGR))

                    print(f"✅ Objet ROND {i} extrait de {filename} → {output_path}")


# === Exemple d'utilisation ===
cleaned_dir = os.path.join(Abra_dir, "round_objects1")
cropped_dir = os.path.join(Abra_dir)
split_round_objects_in_directory(cropped_dir, cleaned_dir)
```

```python
def is_round_diameter(mask_obj, tol=0.15, num_directions=16):
    """
    Vérifie si un objet est rond en mesurant plusieurs diamètres.
    tol : tolérance relative entre diamètres (ex: 0.15 = 15%).
    num_directions : nombre de directions pour mesurer le diamètre.
    """
    contours, _ = cv2.findContours(mask_obj, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return False
    
    cnt = max(contours, key=cv2.contourArea)
    
    # Centre de l'objet
    M = cv2.moments(cnt)
    if M['m00'] == 0:
        return False
    cx = int(M['m10'] / M['m00'])
    cy = int(M['m01'] / M['m00'])

    # Coordonnées des points du contour
    points = cnt[:, 0, :]  # Nx2

    # Calcul des distances radiales dans plusieurs directions
    angles = np.linspace(0, 2 * np.pi, num_directions, endpoint=False)
    diameters = []

    for theta in angles:
        # Direction
        dx = np.cos(theta)
        dy = np.sin(theta)

        # Projeter tous les points du contour sur cette direction
        projections = (points[:, 0] - cx) * dx + (points[:, 1] - cy) * dy
        diam = projections.max() - projections.min()
        diameters.append(diam)

    diameters = np.array(diameters)
    mean_d = diameters.mean()

    # Vérifier que tous les diamètres sont proches du diamètre moyen
    if np.all(np.abs(diameters - mean_d) / mean_d < tol):
        return True
    return False

```

```python
import cv2
import numpy as np
import os

def is_round_diameter(mask_obj, tol=0.15, num_directions=16):
    """
    Vérifie si un objet est rond en mesurant plusieurs diamètres sur le contour.
    tol : tolérance relative entre diamètres (ex: 0.15 = 15%).
    num_directions : nombre de directions pour mesurer le diamètre.
    """
    contours, _ = cv2.findContours(mask_obj, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return False
    
    cnt = max(contours, key=cv2.contourArea)
    if len(cnt) < 5:
        return False  # trop petit pour être fiable

    # Centre de l'objet
    M = cv2.moments(cnt)
    if M['m00'] == 0:
        return False
    cx = M['m10'] / M['m00']
    cy = M['m01'] / M['m00']

    # Coordonnées des points du contour
    points = cnt[:, 0, :].astype(np.float32)  # Nx2

    # Angles pour mesurer diamètres
    angles = np.linspace(0, 2*np.pi, num_directions, endpoint=False)
    directions = np.stack([np.cos(angles), np.sin(angles)], axis=1)  # Dx2

    # Calcul vectorisé des projections
    diffs = points[:, None, :] - np.array([[cx, cy]])  # Nx1x2
    projections = np.sum(diffs * directions[None, :, :], axis=2)  # NxD

    diameters = projections.max(axis=0) - projections.min(axis=0)
    mean_d = diameters.mean()

    return np.all(np.abs(diameters - mean_d) / mean_d < tol)

def split_round_objects_in_directory(input_dir, output_dir, min_area=1000):
    os.makedirs(output_dir, exist_ok=True)

    for filename in os.listdir(input_dir):
        if not filename.lower().endswith(".jpg"):
            continue
        input_path = os.path.join(input_dir, filename)
        image_rgb = cv2.imread(input_path)
        if image_rgb is None:
            continue

        image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)
        _, image_thresh = cv2.threshold(image_gray, 128, 255, cv2.THRESH_BINARY)
        image_inv = cv2.bitwise_not(image_thresh)

        # Supprimer le plus gros contour noir
        contours, _ = cv2.findContours(image_inv, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        mask = np.ones_like(image_thresh) * 255
        if contours:
            biggest = max(contours, key=cv2.contourArea)
            cv2.drawContours(mask, [biggest], -1, 0, thickness=cv2.FILLED)

        # Extraire objets
        num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(mask, connectivity=8)

        for i in range(1, num_labels):
            x, y, w, h, area = stats[i]
            if area < min_area:
                continue

            mask_obj = np.zeros_like(mask)
            mask_obj[labels == i] = 255

            if not is_round_diameter(mask_obj, tol=0.15):
                print(f"❌ Objet {i} de {filename} rejeté (pas rond)")
                continue

            mask_rgb = cv2.cvtColor(mask_obj, cv2.COLOR_GRAY2BGR)
            image_masked = cv2.bitwise_and(image_rgb, mask_rgb)
            image_cropped = image_masked[y:y+h, x:x+w]

            base, ext = os.path.splitext(filename)
            output_path = os.path.join(output_dir, f"{base}_round_{i}{ext}")
            cv2.imwrite(output_path, image_cropped)
            print(f"✅ Objet ROND {i} extrait de {filename} → {output_path}")


# === Exemple d'utilisation ===
Abra_dir = "image/alternaria_brassicicola"
cleaned_dir = os.path.join(Abra_dir, "round_objects3")
split_round_objects_in_directory(Abra_dir, cleaned_dir, tol =0.05)
```

```python
import cv2
import numpy as np
import os

def is_round_vote(mask_obj, tol=0.15, num_direct=10):
    """
    Vérifie si un objet est rond en utilisant des diamètres orthogonaux avec vote majoritaire.
    mask_obj : masque binaire de l'objet
    tol : tolérance relative entre diamètres orthogonaux
    num_direct : nombre de directions aléatoires à tester
    """
    contours, _ = cv2.findContours(mask_obj, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return False
    
    cnt = max(contours, key=cv2.contourArea)
    if len(cnt) < 5:
        return False

    # Centre de l'objet
    M = cv2.moments(cnt)
    if M['m00'] == 0:
        return False
    cx = M['m10'] / M['m00']
    cy = M['m01'] / M['m00']

    # Points du contour
    points = cnt[:, 0, :].astype(np.float32)
    diffs = points - np.array([cx, cy])

    votes = 0
    for _ in range(num_direct):
        # angle aléatoire
        theta = np.random.uniform(0, 2*np.pi)
        dir1 = np.array([np.cos(theta), np.sin(theta)])
        dir2 = np.array([-np.sin(theta), np.cos(theta)])  # orthogonale

        # projections vectorisées
        proj1 = diffs @ dir1
        proj2 = diffs @ dir2
        diam1 = proj1.max() - proj1.min()
        diam2 = proj2.max() - proj2.min()

        # vote
        if np.abs(diam1 - diam2) / max(diam1, diam2) < tol:
            votes += 1

    # vote majoritaire
    return votes > num_direct / 2


def split_round_objects_in_directory(input_dir, output_dir, min_area=1000, tol=0.15, num_direct=10):
    """
    Traite un dossier d'images et extrait les objets ronds avec vote majoritaire sur diamètres orthogonaux.
    min_area : aire minimale pour filtrer les objets
    tol : tolérance relative pour la circularité
    num_direct : nombre de directions aléatoires pour le vote
    """
    os.makedirs(output_dir, exist_ok=True)

    for filename in os.listdir(input_dir):
        if not filename.lower().endswith(".jpg"):
            continue
        input_path = os.path.join(input_dir, filename)
        image_rgb = cv2.imread(input_path)
        if image_rgb is None:
            continue

        # Prétraitement comme dans ta version originale
        image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)
        _, image_thresh = cv2.threshold(image_gray, 128, 255, cv2.THRESH_BINARY)
        image_inv = cv2.bitwise_not(image_thresh)

        contours, _ = cv2.findContours(image_inv, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        mask = np.ones_like(image_thresh) * 255
        if contours:
            biggest = max(contours, key=cv2.contourArea)
            cv2.drawContours(mask, [biggest], -1, 0, thickness=cv2.FILLED)

        # Extraction des objets
        num_labels, labels, stats, _ = cv2.connectedComponentsWithStats(mask, connectivity=8)

        for i in range(1, num_labels):
            x, y, w, h, area = stats[i]
            if area < min_area:
                continue

            mask_obj = np.zeros_like(mask)
            mask_obj[labels == i] = 255

            if not is_round_vote(mask_obj, tol=tol, num_direct=num_direct):
                print(f"❌ Objet {i} de {filename} rejeté (pas rond)")
                continue

            mask_rgb = cv2.cvtColor(mask_obj, cv2.COLOR_GRAY2BGR)
            image_masked = cv2.bitwise_and(image_rgb, mask_rgb)
            image_cropped = image_masked[y:y+h, x:x+w]

            base, ext = os.path.splitext(filename)
            output_path = os.path.join(output_dir, f"{base}_round_{i}{ext}")
            cv2.imwrite(output_path, image_cropped)
            print(f"✅ Objet ROND {i} extrait de {filename} → {output_path}")


# === Exemple d'utilisation ===
Abra_dir = "image/alternaria_brassicicola"
cleaned_dir = os.path.join(Abra_dir, "round_objects_vote_majority")

# Paramètres ajustables : aire min, tolérance, nombre de directions
split_round_objects_in_directory(Abra_dir, cleaned_dir, min_area=1000, tol=0.10, num_direct=12)
```
